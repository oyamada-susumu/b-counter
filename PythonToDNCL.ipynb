```python
#######################################################################S
# Python から DNCL2 に変換します。
# Pythonのソースプログラムを'''と'''の間に張り付けて実行してください。
#######################################################################
pysource='''
def 要素数(x):
    return len(x)
def 文字数(x):
    return len(x)
Tokuten=[8,2,10,6,7,1,9,15,11,3,13,12,4,14,5]
n = len(Tokuten)
saidai = Tokuten[0]
kensu  = 1
print('店番号=',0,'得点=',saidai)
for i in range(1, n, 1):
    if saidai < Tokuten[i]:
        saidai = Tokuten[i]
        kensu = kensu + 1
        print('店番号=',i,'得点=',saidai)
print(kensu)
for j in range(1,10,a):
    print(j)
for k in range(10,-4,-2):
    print(k)
k = 0
while k < 10:
    print(k)
    k = k + 1
while True:
    print(k)
    k = k + 1
x = ''
n = 0
if n > 9:
    x = n
elif n == 10:
    x = 'A'
elif n == 11:
    x = 'B'
elif n == 12:
    x = 'C'
elif n == 13:
    x = 'D'
elif n == 14:
    x = 'E'
elif n == 15:
    x = 'F'
else:
    x = 'X'
print(x)
'''
###############################################################################
# str.isnumeric() str.isdecimal() では負数はTrueにならずFalseとなるので
# 次の関数を使う
def is_num(a):
    try:
        int(a)
    except:
        return False
    return True
################################################################################
# 関数 fuyakuri range(?,?,?) を「増やしながら繰り返す:」に変換する
# text typeA = 'range(5):'           --> arg=['5'] len(arg)=1
# text typeB = 'range(1, n-1):'      --> arg=['1','n-1'] len(arg)=2
# text typeC = 'range(10, -5 , -2):' --> arg=['10','-5','-2'] len(arg)=3
# range(5): 0から4まで1ずつ増やす:　〇
# range(a): 0からa-1まで1ずつ増やす:　〇 ただし a > 0
# range(1,5): 1から4まで1ずつ増やす:　〇
# range(a,b): aからb-1まで1ずつ増やす: △ a < b のとき実行される
# range(a,5): aから4まで1ずつ増やす:　△　a < 5 のとき実行される
# range(1,b): 1からb-1まで1ずつ増やす: △ 1 < b のとき実行される
# range(〇)とrange(〇,〇):は「1ずつ増やす:」だけ
# range(1,5,1): 1から4まで1ずつ増やす:　〇
def fuyakuri(txt):
    txt = txt.replace(' ','')
    txt = txt.replace('range','')
    txt = txt.replace('(','')
    txt = txt.replace(')','')
    txt = txt.replace(':','')
    arg = txt.split(',')
    s = len(arg); n0=0; n1=0; n2=0
    nf0=False; nf1=False; nf2=False
    if s > 0 and is_num(arg[0]) == True:
        nf0 = True; n0 = int(arg[0])
    if s > 1 and is_num(arg[1]) == True:
        nf1 = True; n1 = int(arg[1])
    if s > 2 and is_num(arg[2]) == True:
        nf2 = True; n2 = int(arg[2])
    if s == 1:
        if nf0 == True:
            n0 = n0 - 1
            tx = '0 から '+str(n0)+' まで 1 ずつ増やしながら繰り返す:'
        else:
            tx = '0 から '+arg[0]+'-1 まで 1 ずつ増やしながら繰り返す:'
    elif s == 2:
        if nf1 == True:
            n1 = n1 - 1
            tx = arg[0]+' から '+str(n1)+' まで 1 ずつ増やしながら繰り返す:'
        else:
            tx = arg[0]+' から '+arg[1]+'-1 まで 1 ずつ増やしながら繰り返す:'
    elif s == 3:
        if nf2 == True:
            if n2 > 0:
                if nf1 == True:
                    n1 = n1 - 1
                    tx = arg[0]+' から '+str(n1)+' まで '+str(n2)+' ずつ増やしながら繰り返す:'
                else:
                    tx = arg[0]+' から '+arg[1]+'-1 まで '+str(n2)+' ずつ増やしながら繰り返す:'
            else:
                if nf1 == True:
                    n1 = n1 + 1
                    tx = arg[0]+' から '+str(n1)+' まで '+str(abs(n2))+' ずつ減らしながら繰り返す:'
                else:
                    tx = arg[0]+' から '+arg[1]+'+1 まで '+str(abs(n2))+' ずつ減らしながら繰り返す:'
        else:
            if nf1 == True:
                n1 = n1 - 1
                tx = arg[0]+' から '+str(n1)+' まで '+arg[2]+' ずつ増やしながら繰り返す:'
            else:
                tx = arg[0]+' から '+arg[1]+'-1 まで '+arg[2]+' ずつ増やしながら繰り返す:'
    else:
        tx ='error:rangeの引数'
    return tx
################　△ dictは予約語
# 単純に置換できる単語の辞書
# 予約語elifは予約語ifを含んでいるので「elif」が「elもし」にならないように、elifの変換をifの変換より前に行う。
# 文の先頭の単語を削除するときは,インデントがずれないように以下の処理を行う。 while, for,・・・は削除される単語
#「while△i△<△10」は文の先頭から空白なしで「i△<△10の間」となるように,whileを削除した後に残っている先頭の空白（セパレータとして最低1個以上はある）をすべて削除する。
#「for△i」は文の先頭から空白なしで「iを」となるように,forを削除した後に残っている先頭の空白（セパレータとして最低1個以上はある）をすべて削除する。
# ルール：DictA[?][1]が「''」(その単語が文の先頭にあって削除される）ならば、その単語と次の単語の間の空白をすべて削除する。
jisyoA = [
    ['print','表示する'],
    ['elif','そうでなくてもし'],
    ['if','もし'],
    ['else','そうでなければ'],
    ['while',''],
    ['for',''],
    ['in','を'],
    ['break','繰り返しを抜ける'],
    ['def','関数'],
    ['return','戻り値'],
    ['//','÷'],
    ["'",'"'],
    ['input','【外部からの入力】'],
    ['len','要素数'],
    ]
# ならばの辞書
# 文先頭の単語に応じて文末尾の「:」を「ならば:」他に置き換える。
# 予約語elifは予約語ifを含んでいるので辞書にはifを入れておくだけで、ifとともに「:」を「ならは:」に変換してくれる。
# 逆にelifを入れておくと変換が2回行われ「ならは:ならば：」となってしまう。
jisyoB = [
    ['if',':',' ならば:'],
    ['while',':',' の間繰り返す:'],
    ['def',':',' の定義:'],
    ]
# 制御文のキーワード
jisyoC = ['繰り返す','もし','そうでなければ']
# もし　　ならは:
# │
# ⎿
################################################################################
#  プログラム　本体　ここから
################################################################################
pysrc = pysource.split('\n')
dnsrc =[]
for i in range(len(pysrc)):
    txt = pysrc[i]
    for j in range(len(jisyoB)):
        if txt.find(jisyoB[j][0]) > -1:
            txt = txt.replace(jisyoB[j][1],jisyoB[j][2])
    for j in range(len(jisyoA)):
        f = txt.find(jisyoA[j][0])
        if f > -1:
            txt = txt.replace(jisyoA[j][0],jisyoA[j][1])
            if jisyoA[j][1] == '':           # 単語が削除されたものは、セパレータの空白が残っているので
                txt = ' '*f+txt.lstrip(' ')  # 文頭の空白を削除する(left strip)
    f = txt.find('range')
    if f > -1 :
        txt = txt[:f]+fuyakuri(txt[f:])
    f = txt.find('戻り値')
    if f > -1:
        txt = txt.replace('戻り値','戻り値(')+')'
        txt = ' '*f + txt.replace(' ','')
    f = txt.find('True')
    g = txt.find('の間繰り返す:')
    if f > -1 and g > -1:
        txt = ' '*f+'ずっと繰り返す:'
    dnsrc.append(txt)
#
for i in range(len(dnsrc)):
    if dnsrc[i].find(' ') == 0:
        if dnsrc[i+1].find(' ') != 0:
            print('L',dnsrc[i])
    else:
        print(' ',dnsrc[i])
################################################################################
#  プログラム　本体　ここまで
################################################################################
# for while を削除した後のインデントを保持する　〇
# range(10,-5,-2)は,〇10から-4まで2ずつ減らす(-2の絶対値＝2減らす)
# range(10,-5,b)は, △10から-5までbずつ減らす(bの符号はわからないので混乱する）
# range(10,-5,b)は, 〇10から-5までbずつ増やす(bが負であることを期待して)
# 処理の範囲を示す記号「|」と「L」エル大文字で代用するか？　｜　Ⅼ
```
